## From local to Workbench

Begin by defining a UDF in a local Python environment. Then, upload it to your account's cloud database with the `.to_fused` method. The UDF's name will be that of the function name.

```python
import fused

@fused.udf
def my_new_udf():
  import pandas as pd
  return pd.DataFrame({"b": [123]})
```

```python
# Verify successful execution
fused.run(udf=my_new_udf, engine='local')
```

```python
# Upload the UDF to the cloud
my_new_udf.to_fused()
```

## From Workbench to local

Load a UDF saved in Workbench and run it in a local Python environment.

```python
import fused

loaded_udf = fused.load("plinio@fused.io/my_udf")
loaded_udf
```

```python
fused.run(udf=loaded_udf, engine='local')
```
