# CSV Files



## CSV

```python
import fused


@fused.udf
def udf():

    import pandas as pd

    # Read csv file
    # https://catalog.data.gov/dataset/supply-chain-greenhouse-gas-emission-factors-v1-2-by-naics-6
    df = pd.read_csv('https://pasteur.epa.gov/uploads/10.23719/1528686/SupplyChainGHGEmissionFactors_v1.2_NAICS_CO2e_USD2021.csv')

    return df


df = fused.run(udf=udf, engine="local")
df.head()
```

## TSV

```python
import fused


@fused.udf
def udf():

    import pandas as pd

    # Read census .txt file as csv
    df = pd.read_csv('https://www2.census.gov/programs-surveys/acs/summary_file/2022/table-based-SF/documentation/ACS20225YR_Table_Shells.txt', delimiter='|')

    return df


df = fused.run(udf=udf, engine="local")
df.head()
```
